```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def sinWave(a, x): 
  return np.exp(-a*x) * np.sin(2*np.pi*x)
```

```python
x = np.arange(0, 5, .01)
y = sinWave(.5, x)
y2 = sinWave(1, x)
y3 = sinWave(2, x)
```

```python
plt.plot(x, y, lw=1.5, 
         color='xkcd:dark blue', label = 'A = .5')
plt.plot(x, y2, lw = 2.0,
         color='xkcd:scarlet', linestyle='--', label="A = 1")
plt.plot(x, y3, lw = 2.3, linestyle = '-.', label='A = 2')
plt.legend(title = r"y = $e^{-Ax}\sin (2\pi x)$", framealpha=1, fontsize=10, title_fontsize=11)
plt.ylabel("y")
plt.xlabel("x")
plt.title("Decaying Sine Waves")
plt.grid() 
```
